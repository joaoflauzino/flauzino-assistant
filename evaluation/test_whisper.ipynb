```python
from faster_whisper import WhisperModel
from pathlib import Path
```

```python
home_dir = Path.home()
# No Raspberry Pi 4, usamos o modelo 'tiny' ou 'base' para performance.
# O compute_type="int8" ajuda a economizar muita RAM e CPU.
model_size = "base"
model = WhisperModel(model_size, device="cpu", compute_type="int8")

print("Transcrevendo...")
segments, info = model.transcribe(f"{home_dir}/Downloads/teste-whats-3.opus", beam_size=5, language="pt", vad_filter=True)

print(f"Idioma detectado: {info.language} com probabilidade {info.language_probability:.2f}")

for segment in segments:
    print(f"{segment.text}")
```
